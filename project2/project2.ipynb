# Section 5: Data Science
For this section, I will focus on one question: "Does the Consumer Confidence Index (US) significantly impact Nonfarm Employment (Illinois)?" This question is meaningful because it explores how consumer sentiment may predict changes in employment, which is critical for understanding economic trends.

To answer this question, I will use linear regression to model the relationship between the Consumer Confidence Index (independent variable) and Nonfarm Employment (dependent variable). Linear regression is a simple machine learning model that estimates how much one variable impacts another. By evaluating the model’s coefficients and significance levels, I can determine if a strong relationship exists.

```python
df.corr(numeric_only = True)
```

Since the correlation between the Consumer Confidence Index (US) and Nonfarm Employment (Illinois) is 0.894475, it indicates a strong positive linear relationship. This means that as the Consumer Confidence Index increases, Nonfarm Employment in Illinois tends to increase as well. The high correlation suggests that consumer sentiment might significantly influence employment levels, potentially serving as a leading indicator for regional labor market trends.

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
ind = ["Consumer Confidence Index US"]
dep = "Nonfarm Employment Illinois"
df = df.dropna()
model.fit(df[ind],df[dep])
df["Predict Nonfarm Employment Illinois"] = model.predict(df[ind])
df
```

```python
df["error"] = df["Nonfarm Employment Illinois"] - df["Predict Nonfarm Employment Illinois"]
df
```

```python
df["error"].mean()
```

Since the error of prediction is -1.5158245029548803e-13, the prediction is both too low and too high. Using linear regression, I modeled the relationship between the Consumer Confidence Index (US) as the independent variable and Nonfarm Employment (Illinois) as the dependent variable. The near-zero difference between the actual and predicted values (e.g., -1.5158245029548803e-13) suggests that the model fits the data almost perfectly. This result indicates an exceptionally strong relationship between the two variables, consistent with the previously calculated correlation coefficient of 0.894475.

# Section 6: A Different Data Visualization
For a different type of visualization, I will create a box plot to visualize the distribution of CPI (US Ex Food and Energy) across different quartiles of Gross Value of Natural Gas Production. This will provide insights into how inflation might vary depending on natural gas production levels.

```python
df_year.dropna
df_year.plot.bar(
    x="Year",  
    y="CPI U.S. Ex Food and Energy",  
    title="CPI U.S. Ex Food and Energy from 2005 to 2025",
    xlabel="Year",
    ylabel="CPI U.S. Ex Food and Energy",
    grid = True
    )
```

The bar graph directlt shows that CPI of U.S. Ex Food and Energy increase from 2005 to 2025

```python
df_year.plot.bar(
    x="Year",  
    y="Gross Value Natural Gas Production",  
    title="Gross Value Natural Gas Production from 2005 to 2025",
    xlabel="Year",
    ylabel="Gross Value Natural Gas Production",
    grid = True
    )
```

The bar graph directlt shows How much of gross Value Natural Gas Production increased each year from 2005 to 2025

```python
df['Natural_Gas_Production_Quartile'] = pd.qcut(df['Gross Value Natural Gas Production'], q=4, labels=['Q1', 'Q2', 'Q3', 'Q4'])
graph = df.boxplot(
    column="CPI U.S. Ex Food and Energy",  
    by="Natural_Gas_Production_Quartile",  
    grid=False,
    figsize=(10, 6),
    patch_artist=True
)
```

The box plot reveals how inflation, as measured by the CPI (US Ex Food and Energy), varies across different levels of natural gas production. This helps identify whether higher production levels coincide with significant changes in core inflation.
